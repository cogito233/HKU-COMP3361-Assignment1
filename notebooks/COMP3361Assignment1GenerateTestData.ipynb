```python
class DataFliter(object):
    def __init__(self, data_path, data_name):
        self.data = []
        ans = 0
        with open(data_path+data_name, 'r') as f:
            for line in f:
                self.data.append(line)
    def check_valid(self, s):
        return (' at ' in s) or (' in ' in s) or (' of ' in s) or (' for ' in s) or (' on ' in s) or ('**GW**' in s)
    def __getitem__(self, idx):
        return self.data[idx]
    def __len__(self):
        return len(self.data)
```

```python
import transformers
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-base-uncased')
```

```python
unmasker(" he had a guest @-@ starring role [UNK] the television series the bill [MASK] 2000 .")
```

```python
def generate_data(st):
    st=st.replace("<unk>", "[UNK]")
    li=[]
    while st.find("**GW**")!=-1:
        s = st[:]
        s=s.replace("**GW**", "[MASK]", 1)
        s=s.replace("**GW**", "[UNK]")
        li.append(s)
        st=st.replace("**GW**", "[UNK]", 1)
    return li
```

```python
generate_data(test_data[0])
```

```python
label_num = 0
label_list = ['**GW**', 'at', 'in', 'of', 'for', 'on', 'Ġat', 'Ġin',
              'Ġof', 'Ġfor', 'Ġon',' at', ' in', ' of', ' for', ' on']
label_dict = {'**GW**':0, 'at':1, 'in':2, 'of':3, 'for':4, 'on':5,
                           'Ġat':1, 'Ġin':2, 'Ġof':3, 'Ġfor':4, 'Ġon':5,
                           ' at':1, ' in':2, ' of':3, ' for':4, ' on':5,
                }
preds=[]
confusions = 0
for i in test_data:
    pred=[]
    for s in generate_data(i):
        predict_list=unmasker(s)
        flag = False
        for ele in predict_list:
            if (ele["token_str"] in label_list):
                flag = True
                pred.append(ele["token_str"])
                break
        if (not flag):
            pred.append("<unk>")
            confusions += 1
        label_num += 1
        if label_num % 1000==0:
            print(label_num)
        if label_num==50000:
            break
    preds.append(pred)
    if label_num==50000:
        break
```
